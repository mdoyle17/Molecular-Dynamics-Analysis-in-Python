```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt

labelz=[]
allprlow=[]
allprhigh=[]
all_random_waters = [1566, 962, 1685,247]
for i in all_random_waters:
    labelz.append('random water #%d' %i)
    labelz.append('random water #%d' %i)
    labelz.append('random water #%d' %i)
    print('**********Random water statistics #%d**********' %i)
    all_pr_rand1=[]
    random_water1= i
    radius=15
    OEC='Rand1'
    a = pd.read_pickle("precision_recall_random_water_%d_radius_%d_0.5.pkl" %(random_water1, radius))
    b = pd.read_pickle("precision_recall_random_water_%d_radius_%d_1.0.pkl" %(random_water1, radius))
    c = pd.read_pickle("precision_recall_random_water_%d_radius_%d_1.4.pkl" %(random_water1, radius))
    a['Distance']= 0.5
    b['Distance']= 1.0
    c['Distance']= 1.4
    # cols = df.columns.tolist()
    all_pr_rand1.append(a)
    all_pr_rand1.append(b)
    all_pr_rand1.append(c)
    
    allprlow.append(all_pr_rand1[0].iloc[0])
    allprlow.append(all_pr_rand1[1].iloc[0])
    allprlow.append(all_pr_rand1[2].iloc[0])

    allprhigh.append(all_pr_rand1[0].iloc[-1])
    allprhigh.append(all_pr_rand1[1].iloc[-1])
    allprhigh.append(all_pr_rand1[2].iloc[-1])
    print('Lowest thresh:')
    tab = tabulate([ all_pr_rand1[0].iloc[0],all_pr_rand1[1].iloc[0],all_pr_rand1[2].iloc[0]],headers='keys', tablefmt='fancy_grid')
    print(tab)
    
    print('Highest thresh:')
    tab = tabulate([ all_pr_rand1[0].iloc[-1],all_pr_rand1[1].iloc[-1],all_pr_rand1[2].iloc[-1]],headers='keys', tablefmt='fancy_grid')
    print(tab)
```

```python

import matplotlib.patches as mpatches
import numpy as np

fig= plt.figure(figsize= [14,6])
ax = fig.add_subplot(111)
keyz='f1'
finallabelz=[]
i=0
for x in range(0,len(allprhigh)):
    if allprlow[x]['Distance']==0.5:
        plt.scatter(i,allprlow[x][keyz],color='blue', s=100)

        
    elif allprlow[x]['Distance']==1.0:
        plt.scatter(i,allprlow[x][keyz],color='green', s=100)
 
    else:
        plt.scatter(i,allprlow[x][keyz],color='red', s=100)
        finallabelz.append(labelz[x])
        i=i+1

plt.title('%s' %(keyz), fontsize=25)
ax.set_xticks(np.arange(0,len(finallabelz),1))
ax.set_xticklabels(np.array(finallabelz))
plt.ylabel(keyz,fontsize=20)
plt.xticks( rotation=45, fontsize=20)
plt.yticks(fontsize=20)
blue_patch = mpatches.Patch(color='blue', label='0.5 $\AA$')
green_patch = mpatches.Patch(color='green', label='1.0 $\AA$')
red_patch = mpatches.Patch(color='red', label='1.4 $\AA$')
plt.legend(handles=[blue_patch,green_patch, red_patch],fontsize=20)
plt.tight_layout()
plt.savefig('%s.png' %(keyz))


plt.show()
```
