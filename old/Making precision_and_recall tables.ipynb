```python

import matplotlib.patches as mpatches
import numpy as np

fig= plt.figure(figsize= [14,7])
ax = fig.add_subplot(111)
keyz='recall'
finallabelz=[]
i=0
for x in range(0,len(allprlow)):
    if allprlow[x]['Distance']==0.5:
        plt.scatter(i,allprlow[x][keyz],color='blue', s=100)

        
    elif allprlow[x]['Distance']==1.0:
        plt.scatter(i,allprlow[x][keyz],color='green', s=100)
 
    else:
        plt.scatter(i,allprlow[x][keyz],color='red', s=100)
        finallabelz.append(labelz[x])
        i=i+1

plt.title('%s lowest' %(keyz), fontsize=25)
ax.set_xticks(np.arange(0,len(finallabelz),1))
ax.set_xticklabels(np.array(finallabelz))
plt.ylabel(keyz,fontsize=20)
plt.xticks( rotation=90, fontsize=20)
plt.yticks(fontsize=20)
blue_patch = mpatches.Patch(color='blue', label='0.5 $\AA$')
green_patch = mpatches.Patch(color='green', label='1.0 $\AA$')
red_patch = mpatches.Patch(color='red', label='1.4 $\AA$')
plt.legend(handles=[blue_patch,green_patch, red_patch],fontsize=20)
plt.tight_layout()
plt.savefig('%s at lowest threshold.png' %(keyz))


plt.show()
```
